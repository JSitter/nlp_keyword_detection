```python
URL = 'https://archive.org/stream/alicesadventures19033gut/19033.txt'

r = requests.get(url = URL)
```

```python
print(r.__dir__())
```

```python
print([r.cookies])
```

```python
print(r._content)
```

```python
soup = BeautifulSoup(r._content, 'html.parser')

print(soup.prettify())
```

```python
print(soup.pre)
```

```python
start = 'I--DOWN THE RABBIT-HOLE'
end = 'End of the Project Gutenberg EBook of Alice in Wonderland, by Lewis Carroll'
result = re.search( start + ';(.*)' + end, str(soup.pre))
print(result.group(1))
```
